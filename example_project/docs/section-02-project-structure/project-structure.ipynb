####  Tiedown - Tools for Joining Jupyter Notebooks

# Project Structure<span id="cid2"/>

## I. Tiedown File Types<span id="cid3"/>
This section explains how Tiedown projects can be structured.

### A. Notebooks
There are three types of notebooks (`.ipynb` files) that are used in Tiedown projects. 
1. **Content Notebooks:** Content notebooks are where you write the markdown and code cells for your project. A project generally contains several content notebooks. For example, a book written in Tiedown might have a separate content notebook for each chapter. Content notebooks contain special commands and markup that control how Tiedown builds the output.
2. **Template Notebooks:** Templates are notebooks with cells that are copied into content notebooks. For example, a template can be used to give each content notebook a standardized header or footer.
3. **Output Notebooks:** Output notebooks are created by the Tiedown build process. With the exception of some metadata, an output notebook is just a regular Jupyter notebook.

### B. Supporting Files
Jupyter notebooks often depend on other files, like images, code files, or data files. These files are referred to as *supporting files* in the *Tiedown* documenation.

## II. Folder Structure<span id="cid4"/>
The content and template notebooks are contained in the Tiedown project folder.  See figure 1 for a typical project structure.

####  Figure 1, Typical Project Structue
``` yaml
project-folder:
    - templates:
        - main-template.ipynb
    - content:
        - section-01-introduction:
            - introduction.ipynb
        - section-02-project-structure:
            - project-structure.ipynb
        ...
    - output:
        ...
```

* The `templates` folder contains the project's templates. A project can have zero, one, or more template files.<span id="cid5"/>
* The content notebooks are contained in the `content` folder. 
* Unless otherwise specified, Tiedown will write output notebooks to the `output` subfolder. You can specify a different output folder by passing a path to the `tiedown.Project.build()` method.

## III. Organizing Content<span id="cid6"/>
### A. Flat Structure
You can put all of your content notebooks directly in the `content` folder. Tiedown will order the notebooks by notebook file name when it creates the table of contents (TOC).

### B. Content Subfolders
Tiedown also allows you to organize your content into subfolders, where each subfolder contains related content notebooks and supporting files. For the TOC, Tiedown will order the output notebooks first by subfolder name then by the content notebook file name.

The content subfolders can themselves contain subfolders. Tiedown will copy supporting files and subfolders into the output folder regardless of how deeply nested they are in the folder structure. However *Tiedown will not process content notebooks located more than one level down from the `content` folder.* Make sure your content notebooks are in the `content` folder or a direct subfolder.

### C. Mixed Structure
It's fine to have content notebooks both in subfolders and in the `content` folder. A typical project structure has the introduction, table of contents, and index in the top-level `content` folder, with chapters and their supporting files in their own subfolders.

Previous: [Getting Started<span id="cid2"/>](../section-01-getting-started/getting-started.ipynb) | [Table of Contents](../index.ipynb) | Next: [Tiedown Instructions<span id="cid2"/>](../section-03-tiedown-commands/tiedown-instructions.ipynb)<span id="cid7"/>
