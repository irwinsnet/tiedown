####  Tiedown - Tools for Joining Jupyter Notebooks

# Templates<span id="cid2"/>
Tiedeown templates are easy to use. Only two steps are required.
## I. Creating Blocks in Content Notebooks
To create a block, place a raw cell at the beginning of the block and another raw cell at the end.
* The raw cell at the start of the block must contain the action `block: <block-label>` where &lt;block-label&gt; is the name you assign to the block.
* The raw cell at the end of the block must contain the action `endblock:`.

Content notebooks can have more than one block, but each block should have a unique block-label. [Relative link to section links](../section-07-links/links.ipynb)

## II. Inserting Blocks into Templates<span id="cid3"/>
Insert a block into a template by adding a raw cell with the command `insert: <block-label>`. The raw cell will be replaced with the code and markdown cells from the block in the corresponding content notebook.

## III. Default Template Behavior<span id="cid4"/>
Tiedown's default behavior is to assume there is only one template, the template filename is *main.ipynb*, and the template is located in a subfolder of the project folder named *templates*. Tiedown will insert all blocks into the *main.ipynb* template. All of this behavior is easily modified.

## IV. Changing the Default Template Name and Location<span id="cid5"/>
* Change the location of the template files by assigning a `Pathlib.Path` value to the `Project.template_folder_path` attribute. 
* Change the name of the default template by assigning a string filename to the `Project.default_template` attribute.

## V. Using Multiple Templates<span id="cid6"/>
Create as many templates as you want. Just ensure they are all saved to the location specified by `Project.template_folder_path`. To specify which template should be used with a specific content notebook, place raw cell at the top of the content notebook with a `template: <template-name>` action. The &lt;template-name&gt; parameter should be the filename of the template *without* the file type suffix, e.g., *main*, not *main.ipynb*.

## VI. Using No Template
For notebooks that should be directly converted into an output notebook without being decorated by a template, place the action `template: None` in the top raw cell.
