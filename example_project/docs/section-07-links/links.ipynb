####  Tiedown - Tools for Joining Jupyter Notebooks

# Relative Links<span id="cid2"/>
Manually maintaining internal hyperlinks is tedious, especially if a project consists of several notebooks. Tiedown's mechanism for internal links saves a lot of work. Links just work, even after restructuring projects or renaming files.

## I. Define a Label<span id="cid3"/>


The first step to create a relative link is to create a label in the page *you want to link to*, i.e., the page that is the link's detination.
* To create a page-level link, place a `label` action in the notebook's configuration cell, where `<notebook label>` is an arbitrary text label that you assign to the notebook. 

```yaml
label: <notebook label>
```

* To create a link to a specific markdown cell, place the label action in the destination markdown cell.
```yaml
{% label: <notebook label> %}
```

## II. Insert the Link<span id="cid4"/>

To create the hyperlink, place a `rel_path` insert in the markdown link or HTML anchor tag.
```markdown
[Link Text]({{rel_path: <notebook label>}})
<a href="{{rel_path: <notebook label>}}">Link Text</a>
```
For example: 
```markdown
[Check out this notebook.]({{rel_path: interesting_notebook}})
```
