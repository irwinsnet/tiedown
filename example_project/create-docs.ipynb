# Example Project
The files in the */example_project* folder comprise the *Tiedown* project that was used to build the documentation notebooks in the repository's */docs* folder. Run the code in this notebook to build the project.

## 1. Import Tiedown

```python
import pathlib
import sys

current_notebook_directory = pathlib.Path().cwd()
sys.path.append(str(current_notebook_directory.parent))
import tiedown as td
```

## 2. Create Project
Next, create a new Tiedown project. Pass in a relative path to the folder that contains the source files. This folder is called the *project folder*.

```python
tdp = td.Project(current_notebook_directory, log=False)
print(tdp.project_path)
```

## 3. Build Project
By default, processed notebooks are written to an *output* folder that is a subfolder of the project folder. For this project we'll override the default behavior and put the output notebooks in the repository's */docs* folder.

```python
output_path = current_notebook_directory.parent / "docs"
tdp.build(output_path)
```
