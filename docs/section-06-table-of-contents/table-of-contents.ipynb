## I. Customizing the TOC<span id="cid3"/>


Tiedown's default behavior is to use the text from the first top-level header within a notebook (single hash) for the notebook's entry in the table of contents. Use the `toc_entry` action to change the text that's used for the notebook's entry in the TOC. Place the `toc_entry` action in the notebook's configuration cell. For example:

```
toc_entry: A Customized Entry for the Table of Contents
```

## II. Excluding a Notebook from the TOC<span id="cid4"/>


Place the `toc_excude` action in a notebook's configuration cell to exclude the notebook from the TOC.
