# Tiedown Commands
Tiedown provides special commands that control how a notebook is rendered. These commands can be used in both templates and content notebooks. The commands organized into three categories:

<span id="cid4"/>

{% ignore %}
## I. Actions
Action commands occur in markdown cells and are enclosed in curly braces and percent signs. For example, the following command adds a term and to the project's index.
```
{% index:
    - title(): code %}
```
The action will be removed from the markdown cell during rendering. Think of it like this: action commands take an action, but they don't add any content to the cell in which they occur.

## II. Inserts
Insert commands also occur in markdown cells. They are enclosed in double curly braces. For example, the `rel_path` command will insert a relative path to another notebook into a markdown cell and can be used to create a relative hyperlink:
```
[Test link to Code Blocks]
```
Insert commands are different than action commands because they *insert* content into their enclosing markdown cells.

## III. Raw Commands
Raw commands affect multiple cells or the entire notebook. They are placed in raw notebook cells. For example, placing the following command in a raw cell at the top of a notebook adds a path to the notebook to the project's paths table.
```
path_label: getting_started
```

<span id="cid5"/>

## IV. YAML
Tiedown commands use [YAML](yaml.org) syntax. YAML stands for *YAML Ain't Markup Language*. Tiedown uses the [PyYAML](https://pyyaml.org/) package to parse the commands into Python dictionaries and lists. You can learn more about YAML at [yaml.org](yaml.org).
