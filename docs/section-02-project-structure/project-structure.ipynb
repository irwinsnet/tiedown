# Tiedown - Tools for Joining Jupyter Notebooks

# Project Structure

<span id="cid4"/>

## I. Tiedown File Types

### A. Notebooks
There are three types of notebooks (`.ipynb` files) that are used in Tiedown projects. 
1. **Content Notebooks:** Content notebooks are where you write the markdown and code cells for your project. A project generally contains several content notebooks. For example, a book written in Tiedown might have a separate content notebook for each chapter. Content notebooks contain special commands and markup that control how Tiedown generates output.
2. **Template Notebooks:** Templates are notebooks with cells that are copied into content notebooks. For example, a template can be used to give each content notebook a standardized header or footer.
3. **Output Notebooks:** Output notebooks are created by the Tiedown build process. With the exception of some metadata, an output notebook is just a regular Jupyter notebook.

### B. Supporting Files
Jupyter notebooks often depend on other files, like images, code files, or data files. These files are referred to as **supporting files** in the *Tiedown* documenation.

<span id="cid5"/>

## II. Folder Structure
The content and template notebooks are contained in the Tiedown project folder.  See figure 1 for a typical project structure.

####  Figure 1, Typical Project Structue
``` yaml
project-folder:
    - templates:
        - main-template.ipynb
    - content:
        - section-01-introduction:
            - introduction.ipynb
        - section-02-project-structure:
            - project-structure.ipynb
        ...
    - output:
        ...
```

<span id="cid6"/>

* The `templates` folder contains the project's templates. A project can have zero, one, or more template files.
* The content notebooks are contained in the `content` folder. 
* Unless otherwise specified, Tiedown will write output notebooks to the `output` subfolder. It's easy to specify a different output folder.

<span id="cid7"/>

## III. Organizing Content
### A. Flat Structure
You can put all of your content notebooks directly in the `content` folder. When Tiedown creates a table of contents (TOC), Tiedown will order the notebooks by notebook file name.

### B. Content Subfolders
Tiedown also allows you to organize your content into subfolders, where each subfolder contains related content notebooks and supporting files. For the TOC, Tiedown will order the output notebooks first by subfolder name then by the content notebook file name.

The content subfolders can themselves contain subfolders. Tiedown will copy supporting files and subfolders into the output folder regardless of how deeply nested they are in the folder structure. Hoever Tiedown will not process content notebooks located more than one level down from the `content` folder. Make sure your content notebooks are in the `content` folder or a direct subfolder.

### C. Mixed Structure
It's fine to have content notebooks in subfolders and in the `content` folder. A typical project structure would be to have included the introduction and table of contents in a content notebook in the top-level `content` folder and place subsequent chapters and their supporting files in their own subfolders.

<span id="cid8"/>

[Table of Contents](../index.ipynb)
