# Tiedown - Tools for Joining Jupyter Notebooks

<span id="cid3"/>

## I. Overview
Tiedown is a Python package for managing sets of related, hyperlinked Jupyter or Google Colab notebooks, where each notebook is a section of a larger document. Tiedown provides the following features:
* Templates
* Automatic Section Numbering
* Table of Content Generation
* Term Index
* Relative Hyperlinks

<span id="cid4"/>

## II. History - Why Tiedown Exists
I created my own Python tutorial to support teaching programming and data analysis to high school robotics students. The tutorial consists of about thirty Jupyter notebooks that are hosted on Github. Maintaining that many notebooks required a lot of tedious, time-consumng work. I was manually creating the table of contents and maintaining numerous relative hyperlinks. It seemed like I spent more time checking my section numbers and fixing link errors than I spent actually improving the tutorial content. Keeping track of where specific programming concepts were covered in the tutorial was nearly impossible.

It finally occurred to me that I could create a set of tools in Python to handle the tedious administrative work. The result is a package I call *Tiedown*. The package helps to *tie* different notebooks together by providing commands that can be included in mark*down* cells. *Tiedown*. Get it?

<span id="cid5"/>

## III. Package Design
### A. Python
*Tiedown* is written in Python mainly because Python is my strongest programming language. I suppose I could have written Tiedown as a JavaScript Jupyter Lab extension, but then a working Jupyter installation would be required to build a project. Jupyter is only one of several different environments that can run Jupyter notebooks (e.g., Visual Studio Code, Google Colab). As written, Tiedown does not require a running Jupyter server to build the output notebooks.

### B. YAGNI
There are countless, helpful features that seem like they should be included in Tiedown. Most of these features are missing. I only included features that I needed for my robotics Python tutorial.

### C. Dependencies
*Tiedown* uses [PyYAML](https://pyyaml.org/wiki/PyYAML) to parse commands and [nbformat](https://nbformat.readthedocs.io/en/latest/index.html) to manage `.ibynb` files.  All other imports are from the [Python Standard Library](https://docs.python.org/3/library/index.html)

### D. Inspiration
Tiedown's templating system and syntax were inspired by the [Jinja templating engine](https://jinja.palletsprojects.com/en/3.1.x/).
