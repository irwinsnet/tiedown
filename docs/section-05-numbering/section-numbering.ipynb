####  Tiedown - Tools for Joining Jupyter Notebooks

# Section Numbering<span id="cid2"/>
Tiedown looks for an `outline` action in every template's command cell. Tiedown will automatically number all markdown header lines if the `outline` command is present and it has a proper outline level argument.

## I. Outline Level Argument<span id="cid3"/>
The outline level arguement used for these documents is `.I.A`.
* The starting period tells Tiedown *not* to apply numbers to level 1 headers (headers that start with one hash character (`#`). This documentation only used level-1 headers for page titles.
* Level-2 headers (`##`) will be numbered with uppercase Roman numerals.
* Level-3 headers (`###`) will be numbered with uppercase letters.
* Level-4 through Level-6 headers will be unumbered.

### A. Number Format Specifiers
Tiedown provides six options for section numbering. Each option has a single-character format specifier.
* `I`: Uppercase Roman Numerals
* `i`: Lowercase Roman Numerals
* `A`: Uppercase Letters
* `a`: Lowercase Letters
* `1`: Ordinary Numbers
* `_` or empty string: No Numbers

The last option is to not number the sections and is specified by an empty string.

The level argument starts with the format specifier for level-1 headers, followed by the format specifier for level-2 headers, and so forth. Each level is separated by a period. If the level argument contains fewer than six format specifiers, then the lower-level headers without a format specifier will be unumbered.

## II. Examples<span id="cid4"/>
* `outline: A.1.i._.I`: Level-1, -2, and -3 heaers will be numered with uppercase letters, numbers, and lowercase Roman numerals. Level-4 headers wil be un-numbered, and level-5 headers will have uppercase Roman numerals. Note that the underscore `_` can be replaced with an empty string without changing the meaning of the level argument. That is, `A.1.i..I` is the same as `A.1.i._.I`, which is the same as `A.1.i._.I._`. 
* `outline: _.1` numbers only the level-2 headers. `.1` is equivalent.

## III. Overriding the Template's `outline` Command<span id="cid5"/>
The template's `outline` action can be overridden by placing an `outline` action in the content notebook's command cell.

## IV. Disabling Numbering<span id="cid6"/>
Use `outline: None` to disable automatic numbering.
