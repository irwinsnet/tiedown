####  Tiedown - Tools for Joining Jupyter Notebooks

# Relative Links<span id="cid2"/>
Manually maintaining internal hyperlinks is tedious, especially if a project consists of several notebooks. Tiedown's mechanism for internal links saves a lot of work. Links just work, even after restructuring projects or renaming files.

## I. Page Links<span id="cid3"/>


There are two steps for creating a Tiedown link:
1. Assign a label to the notebook with the `label` action. The `label` action should go in the notebook's configuration cell. `<notebook label>` is an arbitrary text label that you assign to the notebook. 
```
label: <notebook label>
```
2. Use the `{{ rel_path: <notebook label> }}` insert to insert a relative path to that notebook into a markdown or HTML link. For example:
```markdown
[Check out this notebook.]({{rel_path: interesting_notebook}})
```

## II. Section Links<span id="cid4"/>



Place the `toc_excude` action in a notebook's configuration cell to exclude the notebook from the TOC.
